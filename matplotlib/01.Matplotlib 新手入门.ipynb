```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

import matplotlib.pyplot as plt
# 解决中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']
# plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']  # mac 用这行
# 负号正常显示
plt.rcParams['axes.unicode_minus'] = False
```

###  准备画布
    plt.figure(num=None, figsize=None, dpi=None, facecolor=None)
        num = int  or str , 大画布的标记
        figsize = （x，y）  配合 dpi 
        dpi = 72  ,每个 x，y 对应 72像素
        facecolor = 颜色单词  或  16进制颜色代码  或 简写的颜色

```python
# 准备画图的数据
x = range(1,10)
y = np.random.randint(1,20,9)
```

### 折线图
    plt.plot(x,y,....)

```python
# 准备画布
plt.figure( 1 , figsize=(8,6) ,dpi=72 ,facecolor='pink' )

# 画折线图
plt.plot(x,y , label='图例')

# 显示图例
plt.legend()
# 设置 x ，y 轴的文本
plt.xlabel('我是x轴文本')
plt.ylabel('我是y轴文本')
# 设置 x ，y 轴的数值范围
plt.xlim(0,10) # 0-10
plt.ylim(0,20) # 0-20
# 设置 x ，y 轴的刻度   plt.xticks( 需要显示的刻度list   [,每个刻度对应的文本list]  )
plt.xticks( range(0,11) ,list('零一二三四五六七八九十'))
plt.yticks( range(1,21,2) )
# 设置标题
plt.title('图像的标题title')
# 设置图像的参考线  ,  plt.grid( axis= 'x or y  or  both' )  ,默认 both
plt.grid(axis='y')

# 只显示图像
plt.show()
```

###  柱状图
    plt.bar( x,y, ....)

```python
# 准备画布
plt.figure( 1 , figsize=(8,6) ,dpi=72 ,facecolor='pink' )

# 画柱状图
plt.bar(x,y , label='图例')

# 显示图例
plt.legend()
# 设置 x ，y 轴的文本
plt.xlabel('我是x轴文本')
plt.ylabel('我是y轴文本')
# 设置 x ，y 轴的数值范围
plt.xlim(0,10) # 0-10
plt.ylim(0,20) # 0-20
# 设置 x ，y 轴的刻度   plt.xticks( 需要显示的刻度list   [,每个刻度对应的文本list]  )
plt.xticks( range(0,11) ,list('零一二三四五六七八九十'))
plt.yticks( range(1,21,2) )
# 设置标题
plt.title('图像的标题title')
# 设置图像的参考线  ,  plt.grid( axis= 'x or y  or  both' )  ,默认 both
plt.grid(axis='y')

# 只显示图像
plt.show()
```

####  可以画在一起

```python
# 准备画布
plt.figure( 1 , figsize=(8,6) ,dpi=72 ,facecolor='pink' )

# 画折线图 + 柱状图
plt.plot(x,y , label='折线图',color='red')
plt.bar(x,y , label='柱状图')

# 显示图例
plt.legend()
# 设置 x ，y 轴的文本
plt.xlabel('我是x轴文本')
plt.ylabel('我是y轴文本')
# 设置 x ，y 轴的数值范围
plt.xlim(0,10) # 0-10
plt.ylim(0,20) # 0-20
# 设置 x ，y 轴的刻度   plt.xticks( 需要显示的刻度list   [,每个刻度对应的文本list]  )
plt.xticks( range(0,11) ,list('零一二三四五六七八九十'))
plt.yticks( range(1,21,2) )
# 设置标题
plt.title('图像的标题title')
# 设置图像的参考线  ,  plt.grid( axis= 'x or y  or  both' )  ,默认 both
plt.grid(axis='y')

# 只显示图像
plt.show()
```

### 多个直线图

```python
plt.figure()

x = range(10)
y1 = np.random.uniform(0,20,10)
y2 = np.random.uniform(-10,10,10)

# 同一种图，多个会自动分配颜色
plt.plot(x,y1 ,label='y1')
plt.plot(x,y2 ,label='y2')

plt.legend()

plt.show()
```
