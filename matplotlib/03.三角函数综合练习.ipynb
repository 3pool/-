#### 3、控制轴
    ax = plt.gca() 
    ax.spines  获取四个轴对象

```python
# 准备画布
plt.figure(figsize=(10,6))

# 准备 x ，cos ，sin
x = np.linspace(-np.pi , np.pi ,256)
ysin = np.sin(x)
ycos = np.cos(x)
te = np.pi*2/3

# 折线图
plt.plot(x,ysin ,color='red',linewidth=2.5 ,label='sin')
plt.plot(x,ycos ,color='blue',lw=2.5 ,label='cos')

# 虚线连接
plt.plot( [te,te],[0,np.sin(te)] ,color='red',linestyle='--')
plt.plot( [te,te],[0,np.cos(te)] ,color='blue',ls='--')

# 两个散点
plt.scatter([te],[np.sin(te)] ,color='red',s=60)
plt.scatter([te],[np.cos(te)] ,color='blue',s=60)

plt.legend(loc='upper left')

# 显示刻度与值
pi = np.pi
plt.xticks( [-pi ,-pi/2 ,0 ,pi/2,pi] ,['-π',r'$-\frac{\pi}{2}$','0',r'$\frac{\pi}{2}$','π'],
          fontfamily='monospace',fontsize=14
          )
plt.yticks([-1,1])

# 3、设置轴
ax = plt.gca()
# 让右边 与 上面的颜色消失
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# 让下面与左边到 数据中心  ('data',0)
ax.spines['bottom'].set_position( ('data',0) )
ax.spines['left'].set_position( ('data',0) )

plt.show()
```

#### 4、标记点位
    plt.text(x,y,str)
    plt.annotate(text , xy ,...)

```python
# 准备画布
plt.figure(figsize=(10,6))

# 准备 x ，cos ，sin
x = np.linspace(-np.pi , np.pi ,256)
ysin = np.sin(x)
ycos = np.cos(x)
te = np.pi*2/3

# 折线图
plt.plot(x,ysin ,color='red',linewidth=2.5 ,label='sin')
plt.plot(x,ycos ,color='blue',lw=2.5 ,label='cos')

# 虚线连接
plt.plot( [te,te],[0,np.sin(te)] ,color='red',linestyle='--')
plt.plot( [te,te],[0,np.cos(te)] ,color='blue',ls='--')

# 两个散点
plt.scatter([te],[np.sin(te)] ,color='red',s=60)
plt.scatter([te],[np.cos(te)] ,color='blue',s=60)

plt.legend(loc='upper left')

# 显示刻度与值
pi = np.pi
plt.xticks( [-pi ,-pi/2 ,0 ,pi/2,pi] ,['-π',r'$-\frac{\pi}{2}$','0',r'$\frac{\pi}{2}$','π'],
          fontfamily='monospace',fontsize=14
          )
plt.yticks([-1,1])

# 3、设置轴
ax = plt.gca()
# 让右边 与 上面的颜色消失
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# 让下面与左边到 数据中心  ('data',0)
ax.spines['bottom'].set_position( ('data',0) )
ax.spines['left'].set_position( ('data',0) )

# 4、标记文本
# sin
plt.annotate(
    r'$sin(\frac{2π}{3}) = \frac{\sqrt{3}}{2}$',
    xy = (te,np.sin(te)),
    fontsize=16,
    xycoords='data', textcoords = 'offset points',xytext=(20,25),
    arrowprops = {'arrowstyle':'->','connectionstyle':'arc3,rad=0.5'}
)
# cos
plt.annotate(
    r'$cos(\frac{2π}{3}) = \frac{1}{2}$',
    xy = (te,np.cos(te)),
    fontsize=16,
    xycoords='data', textcoords = 'offset points',xytext=(-140,-40),
    arrowprops = {'arrowstyle':'->','connectionstyle':'arc3,rad=0.5'}
)


plt.show()
```
