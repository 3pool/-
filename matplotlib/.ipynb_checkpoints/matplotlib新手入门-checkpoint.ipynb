```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
#解决中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']
#plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']  #mac用该字体
#负号正常显示
plt.rcParams['axes.unicode_minus'] = False
```

```python
x = range(1,10)
y = np.random.randint(1,20,9)
```

```python
#准备画布
plt.figure(1,figsize=(8,6),dpi=72,facecolor='pink')
```

```python
#数据代入
plt.plot(x,y)
```

```python
#显示图例
plt.plot(x,y,label='图例')
plt.legend()
#设置轴标题
plt.xlabel('我是x轴文本')
plt.ylabel('我是y轴文本')
#设置轴范围边界
plt.xlim(0,10)
plt.ylim(0,20)
#设置刻度长度和名称
plt.xticks(range(0,11),list('零一二三四五六七八九十'))
plt.yticks(range(1,21,2))
#设置标题
plt.title('图像的标题')
#设置参考线
#plt.grid(axis='x')
plt.grid(axis='y')
#plt.grid(axis='both')
#只显示图像
plt.show()
```

```python
#准备画布
plt.figure(1,figsize=(8,6),dpi=72,facecolor='pink')
#显示图例
plt.bar(x,y,label='图例')
plt.legend()
#设置轴标题
plt.xlabel('我是x轴文本')
plt.ylabel('我是y轴文本')
#设置轴范围边界
plt.xlim(0,10)
plt.ylim(0,20)
#设置刻度长度和名称
plt.xticks(range(0,11),list('零一二三四五六七八九十'))
plt.yticks(range(1,21,2))
#设置标题
plt.title('图像的标题')
#设置参考线
#plt.grid(axis='x')
plt.grid(axis='y')
#plt.grid(axis='both')
#只显示图像
plt.show()
```

```python
#准备画布
plt.figure(1,figsize=(8,6),dpi=72,facecolor='pink')
#显示图例
plt.plot(x,y,label='折线图',color='red')
plt.bar(x,y,label='柱状图')
plt.legend()
#设置轴标题
plt.xlabel('我是x轴文本')
plt.ylabel('我是y轴文本')
#设置轴范围边界
plt.xlim(0,10)
plt.ylim(0,20)
#设置刻度长度和名称
plt.xticks(range(0,11),list('零一二三四五六七八九十'))
plt.yticks(range(1,21,2))
#设置标题
plt.title('图像的标题')
#设置参考线
#plt.grid(axis='x')
plt.grid(axis='y')
#plt.grid(axis='both')
#只显示图像
plt.show()
```
