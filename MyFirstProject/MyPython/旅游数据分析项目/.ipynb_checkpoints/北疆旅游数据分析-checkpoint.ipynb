```python
import csv
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
data = pd.read_csv('trip.csv')
```

```python
data.duplicated().value_counts()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.info()
```

```python
data['number'].fillna('0',inplace=True)
```

```python
data['transport'].fillna('飞机往返',inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
data['number'] = data['number'].astype('int64')
```

```python
data.reset_index(drop=True,inplace=True)
```

```python
data.info()
```

```python
data['departure'] = data['departure'].apply(lambda x:x[:-2])
```

```python
data['departure'].value_counts().head()
```

```python
data['departure'] = data['departure'].apply(lambda x:x.replace('/香港',''))
```

```python
data['departure'].value_counts().head()
```

```python
data['mode'].value_counts()
```

```python
data['transport'].value_counts()
```

```python
data['agency'].value_counts().head()
```

```python
data.describe()
```

```python
data['isOptimization'].value_counts()
```

```python
data.head()
```

```python
allplace = list(pd.read_csv('景点.txt'))
```

```python
allplace
```

```python
def get_place(x):
    place=[]
    for each in allplace:
        if each in x:
            place.append(each)
        else:
            pass
    return place
```

```python
goplace = data.title.apply(get_place)
```

```python
goplace
```

```python
data['attraction'] = goplace
```

```python
data
```

```python
night = data['title'].apply(lambda x : x.split('晚')[0].split('日')[-1]).astype(int)
```

```python
day = night+1
```

```python
data['tripDate'] = day.astype(str)+'日'+night.astype(str)+'晚'
```

```python
data['newTags'] = data.tags.apply(lambda x : x.split('/'))
```

```python
data.head(10)
```

```python
data.drop(['title','tags'],axis=1,inplace=True)
```

```python
data.head()
```

```python
from pyecharts import *
```

```python
mode_X = data['mode'].value_counts().index
```

```python
mode_A = data['mode'].value_counts().values
```

```python
pie = Pie('出团类型整体占比情况',title_pos='center')
pie.add('出团类型',mode_X,mode_A,
       is_label_show=True,
       label_pos='center',
       legend_orient="vertical",
       legend_pos="600",
       radius = [40,75])
```

```python
mode_BJ = data[data['departure']=='北京']['mode'].value_counts()
mode_SH = data[data['departure']=='上海']['mode'].value_counts()
mode_HZ = data[data['departure']=='杭州']['mode'].value_counts()
mode_NJ = data[data['departure']=='南京']['mode'].value_counts()
```

```python
pie = Pie('出团类型部分城市占比情况',title_pos='center')
pie.add('出团类型-北京',mode_BJ.index,mode_BJ.values,
        center=[40,30] ,radius=[22,30],label_pos='center',legend_orient="vertical",legend_pos="600",is_label_show=True)
pie.add('出团类型-上海',mode_SH.index,mode_SH.values,
        center=[60,30] ,radius=[22,30],label_pos='center',legend_orient="vertical",legend_pos="600",is_label_show=True)
pie.add('出团类型-杭州',mode_HZ.index,mode_HZ.values,
        center=[40,70] ,radius=[22,30],label_pos='center',legend_orient="vertical",legend_pos="600",is_label_show=True)
pie.add('出团类型-南京',mode_NJ.index,mode_NJ.values,
        center=[60,70] ,radius=[22,30],label_pos='center',legend_orient="vertical",legend_pos="600",is_label_show=True)
```

```python
transport_X = data['transport'].value_counts().index
```

```python
transport_A = data['transport'].value_counts().values
```

```python
bar = Bar('出行交通方式对比情况',title_pos = 'center')
bar.add('',transport_X,transport_A,is_label_show=True,bar_category_gap='50%')
```

```python
agency_X = data['agency'].value_counts().head(5).index
agency_A = data['agency'].value_counts().head(5).values
```

```python
bar = Bar('各供应商提供的产品数量情况',title_pos = 'center')
bar.add('',agency_X,agency_A,is_label_show=True,
        xaxis_rotate = 8,bar_category_gap='50%')
```

```python
isOptimization_X = data['isOptimization'].value_counts().index.sort_values()
isOptimization_A = data.groupby('isOptimization')['number'].count().values
isOptimization_B = data.groupby('isOptimization')['number'].sum().values
```

```python
bar = Bar('是否优选与销量的关联情况',title_pos = 'center')
bar.add('产品数量',isOptimization_X,isOptimization_A,
        is_label_show=True)
bar.add('产品销量',isOptimization_X,isOptimization_B,
        is_label_show=True,legend_pos="600")
```

```python
tripDate_X = data.groupby('tripDate').number.sum().sort_values(ascending=False).index
tripDate_A = data.groupby('tripDate').number.sum().sort_values(ascending=False).values
```

```python
bar = Bar('不同出行时长的受欢迎程度情况',title_pos = 'center')
bar.add('',tripDate_X,tripDate_A,is_label_show=True,
        xaxis_rotate = 30,bar_category_gap='50%')
```

```python
scatter = Scatter("价格与销量的关联情况",title_pos='center')
scatter.add("", data.price,data.number)
```

```python
geo = Geo(
    "各出发城市价格差异情况",
    "",
    title_color="#fff",
    title_pos="center",
    width=600,
    height=400,
    background_color="#404a59",
)
geo.add(
    "",
    data.groupby('departure').price.mean().index,
    data.groupby('departure').price.mean().values,
    type="scatter",
    visual_range=[7253,12380],
    visual_text_color="#fff",
    symbol_size=15,
    is_visualmap=True,
)
```

```python
attraction_X = pd.value_counts(np.concatenate(list(data.attraction))).sort_values().index
attraction_A = pd.value_counts(np.concatenate(list(data.attraction))).sort_values().values
```

```python
bar = Bar('各景点在所有产品中出现的频率情况',title_pos = 'center')
bar.add('',attraction_X,attraction_A,
       is_convert=True,is_datazoom_show=True,
        datazoom_orient='vertical'
       )
bar.render()
```
