```python
import pandas as pd
TrainData = pd.read_csv("./datasets/train_processed.csv")
TrainData.head()
```

```python
TrainData.shape
```

```python
# trainData：为所有训练数据集，trainData_X:训练集自变量(x)，trainData_y:训练集因变量(y)
# 拆分出特征变量（自变量）
trainData_X = TrainData.drop(['Survived'], axis = 1)
# 拆分出目标变量（因变量）
trainData_y = TrainData.Survived 
```

```python
# 从sklearn中导入数据集拆分方法 train_test_split
from sklearn.model_selection import train_test_split

# X_test:代表的是测试集数据（自变量），y_test:代表的是测试集数据（因变量）
train_x, test_x, train_y, test_y = train_test_split(trainData_X, # 数据集的特征变量
                                                    trainData_y, # 数据集的目标变量
                                                    test_size=0.3, # 测试集占比30%
                                                    random_state=123456) # 用于将样本顺序打乱的一个随机数
# train_x 训练集的特征变量
# train_y 训练集的目标变量
# test_x 测试及的特征变量
# test_y 测试集的目标变量
```

```python
# 查阅KFold相关的资料，K折交叉验证
```

```python
# 训练数据集的特征变量
train_x.shape
```

```python
test_x.shape
```

```python
# 导入逻辑回归算法，注意：这是一个分类算法
from sklearn.linear_model import LogisticRegression

# 模型实例化（未经训练的模型，等待需要输入训练数据训练）
lr = LogisticRegression()

# 训练模型（训练数据输入到算法中进行训练）
lr.fit(train_x, train_y)
```

```python
# 从sklearn中导入分类模型评估报告模块 classification_report
from sklearn.metrics import classification_report

# 用模型对测试集进行预测（用之前得到的y=f(x)公式，带入x,求解y）
y_test_pred = lr.predict(test_x)


# 输出测试集上模型评估报告classification_report(即预测试的结果集和实际事先标注的真实结果集进行比较)
print(classification_report(test_y,y_test_pred))
```

```python
# 去除掉PassengerId，将特征变量Test_X从数据集筛选出来
Test_X = TestData.drop(['PassengerId'], axis = 1)

# 使用模型对新的数据进行预测
pred = lr.predict(Test_X)

# 把PassengerId和pred通过DataFrame()拼接起来形成最后的提交结果
Submission = pd.DataFrame({'PassengerId':TestData.PassengerId,'Survived':pred})
Submission.head()
```

```python
# 输出部分预测类别
print(pred[:10])
```

```python
# 输出样本预测类别概率
y_test_pred_prob = lr.predict_proba(Test_X)
# 值显示前10行
y_test_pred_prob[:10]
```

**以上概率输出中，第一列为乘客遇难的概率 $P(y=0)$，第二列是乘客生还的概率$P(y=1)$**

```python
# 类别预测结果,封装成DataFrame
pred_df = pd.DataFrame(pred,columns=['pred'])

# 概率预测结果，封装成DataFrame
pred_prob_df = pd.DataFrame(y_test_pred_prob,columns=['p_0','p_1'])

# 拼接乘客Id、预测类别以及类别的概率
pd.concat((Submission[["PassengerId"]],
           pred_df,
           pred_prob_df),
          axis=1).head()  # axis=1表示水平方向拼接
```

```python
# 输出模型的截距
print(lr.intercept_ )
```

```python
# 输出模型的回归系数
print(lr.coef_)
```

```python
lr.coef_.shape # 40个回归系数对应着40个特征变量
```

```python
# 将回归系数与特征变量拼接
s = "y = "
for index,coef in enumerate(lr.coef_.flatten().tolist()):
#     print(index,coef)
    if coef < 0:
        s += "{:.2}*x{}".format(coef,index+1)
    else:
        s += "+{:.2}*x{}".format(coef,index+1)
# 将截距b拼接到方程中
s += str(lr.intercept_[0])
print(s)
```

```python
# 将回归系数与相应特征组合成DataFrame
import numpy as np
# 把变量名称和系数对应起来组合成dataframe
coef = pd.DataFrame(list(zip(train_x.columns, 
                             np.transpose(lr.coef_))), 
                    columns=['columns', 'coef'])
# 按照coef的值从高到底排序
coef = coef.sort_values(by =['coef'], ascending=False)
coef
```
